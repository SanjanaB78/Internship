```python
X=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.predict([[7.6]])
```

```python
y_p=lr.predict(X)
y_p
```

```python
y
```

```python
plt.scatter(X,y)
plt.show()
```

```python
plt.scatter(X,y,color='red',label="Actual Values")
plt.plot(X,y_p,color='k',label="Predicted Values")
plt.xlabel("Amount",fontsize=14)
plt.ylabel("Class",fontsize=14)
plt.title("Linear Regression to predict credit card fraud detedction")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y,y_p)
```

```python
mean_squared_error(y,y_p)
```
